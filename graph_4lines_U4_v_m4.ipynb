```python
import numpy as np
import matplotlib.pyplot as plt

def compute_U4(m, SD_TD): # lower solution

    Qec = 8.61897e5
    Md = 6.53588e9
    b = -1.57262e-4
    a = 1.37626e-7

    A = (2 * (Qec + Md))** 2

    B = ((m**4) + (b * (m**2))) ** 2

    C = 4 * (m**4) * (a - (SD_TD**2)) * A

    D = np.sqrt(C + B)

    E = -1 * ((m**4) + (b * (m**2)))
    
    F = np.sqrt((E + D) / (-2 * (m**4)))

    return F

def f():
    m4 = np.linspace(0.8e5, 7e5, 1000000)
    u41 = compute_U4(m4, 2.9)
    u42 = compute_U4(m4, 2)
    u43 = compute_U4(m4, 1.5)
    u44 = compute_U4(m4, 1)
    plt.plot(m4, u41, label='SD(TD) = 2.9 eV')
    plt.plot(m4, u42, label='SD(TD) = 2 eV')
    plt.plot(m4, u43, label='SD(TD) = 1.5 eV')
    plt.plot(m4, u44, label='SD(TD) = 1 eV')
    plt.xlabel("m_4 (eV)")
    plt.ylabel("U_4")
    plt.title("U_4 vs. m_4")
    plt.legend()
    plt.savefig("graph_4lines_U4_v_m4.png")

def f1():    
    m4 = np.linspace(5e5, 1e6, 1000000)
    u41 = compute_U4(m4, 2.9)
    u42 = compute_U4(m4, 2)
    u43 = compute_U4(m4, 1.5)
    u44 = compute_U4(m4, 1)
    plt.plot(m4, u41, label='SD(TD) = 2.9 eV')
    plt.plot(m4, u42, label='SD(TD) = 2 eV')
    plt.plot(m4, u43, label='SD(TD) = 1.5 eV')
    plt.plot(m4, u44, label='SD(TD) = 1 eV')
    plt.xlabel("m_4 (eV)")
    plt.ylabel("U_4")
    plt.title("U_4 vs. m_4")
    plt.legend()
    plt.savefig("graph_4lines_U4_v_m4_1.png")

def f2():    
    m4 = np.linspace(1e6, 10e6, 1000000)
    u41 = compute_U4(m4, 2.9)
    u42 = compute_U4(m4, 2)
    u43 = compute_U4(m4, 1.5)
    u44 = compute_U4(m4, 1)
    plt.plot(m4, u41, label='SD(TD) = 2.9 eV')
    plt.plot(m4, u42, label='SD(TD) = 2 eV')
    plt.plot(m4, u43, label='SD(TD) = 1.5 eV')
    plt.plot(m4, u44, label='SD(TD) = 1 eV')
    plt.xlabel("m_4 (eV)")
    plt.ylabel("U_4")
    plt.title("U_4 vs. m_4")
    plt.legend()
    plt.savefig("graph_4lines_U4_v_m4_2.png")

def f3():    
    m4 = np.linspace(10e6, 20e6, 1000000)
    u41 = compute_U4(m4, 2.9)
    u42 = compute_U4(m4, 2)
    u43 = compute_U4(m4, 1.5)
    u44 = compute_U4(m4, 1)
    plt.plot(m4, u41, label='SD(TD) = 2.9 eV')
    plt.plot(m4, u42, label='SD(TD) = 2 eV')
    plt.plot(m4, u43, label='SD(TD) = 1.5 eV')
    plt.plot(m4, u44, label='SD(TD) = 1 eV')
    plt.xlabel("m_4 (eV)")
    plt.ylabel("U_4")
    plt.title("U_4 vs. m_4")
    plt.legend()
    plt.savefig("graph_4lines_U4_v_m4_3.png")

def f4():    
    m4 = np.linspace(1e6, 10e6, 1000000)
    u41 = compute_U4(m4, 2.9)
    u44 = compute_U4(m4, 1)
    plt.plot(m4, u41, label='SD(TD) = 2.9 eV')
    plt.plot(m4, u44, label='SD(TD) = 1 eV')
    plt.xlabel("m_4 (eV)")
    plt.ylabel("U_4")
    plt.title("U_4 vs. m_4")
    plt.legend()
    plt.savefig("graph_4lines_U4_v_m4_comparison.png")


f4()



    
```
