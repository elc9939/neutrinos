```python
import numpy as np
import matplotlib.pyplot as plt

def Mc(SD): 
    Qec = 8.61897e5
    Md = 6.53588e9
    c = 1
    b = -1.57263e-4
    a = 1.37626e-7
    
    S = 4 * (a - (SD ** 2)) * ((2 * (Qec + (Md * (c ** 2))) / (c ** 4)) ** 2)
    return np.sqrt(np.sqrt(-1 * S) - b)

x = np.linspace(0.1, 2.9, 1000)
y = Mc(x)

plt.plot(x, y)
plt.title("standard deviation of the recoil energy vs. critical mass")
plt.xlabel("SD(TD) (eV)")
plt.ylabel("m_c (eV)")
plt.subplots_adjust(left=0.15)
plt.savefig("SD_VS_MC.png")
```
