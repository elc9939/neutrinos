```python
import numpy as np
import matplotlib.pyplot as plt

m4_vals = np.logspace(-40, 40, num=1000)
u4max_vals = []

for m in m4_vals:
    f = np.polynomial.Polynomial([-1.437e21, 0, m**4 - 1.573e-4 * m**2, 0,-1 * m**4])
    derv = f.deriv()
    roots = f.roots()
    roots = roots[np.isreal(roots)].real # filter only real
    roots = roots[(roots >= 0) & (roots < 1)] # filter only between 0, 1
    if len(roots) > 0:
        if derv(np.max(roots)) < 0:
            u4max_vals.append(1) # negative derivative means values up to 1 work
            print("1", roots)
        else:
            u4max_vals.append(np.max(roots)) # positive derivative means that the largest value that would work
            print("2")
    else:
        if f(0.5) < 0:
            u4max_vals.append(1) # valid everywhere
            print("3", roots)
        else:
            u4max_vals.append(0) # valid nowhere 
            print("4")

plt.loglog(m4_vals, u4max_vals)
plt.xlabel("m4 (eV)")
plt.ylabel("u4 max")
plt.title("u4 max vs. m4")
plt.legend()
plt.show
```
