```python
import numpy as np
import matplotlib.pyplot as plt

def true_P4(m):
    a = m**4
    b = 1.57e-4*(m**2) - (m**4)
    c = 1.44e21
    return (-1 * b - np.sqrt(b**2  - 4*a*c)) / (2 * a)

def approx_P4(m):
    return 1.44e21 / (m**4)

m4 = np.linspace(1e7, 1e8)
true = true_P4(m4)
approx = approx_P4(m4)

plt.plot(m4, true)
plt.plot(m4, approx)
plt.show()
```
